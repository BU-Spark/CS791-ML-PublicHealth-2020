```python
# %load label_test.py
import pandas as pd

df_enforcement = pd.read_excel("Data.xlsx")
df_compliant = pd.read_excel("FLDExtract.xlsx")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,9))

sns.set()
industry_count = sns.countplot(x="Industry",data=df_enforcement)
industry_count.set_xticklabels(industry_count.get_xticklabels(), rotation=60, horizontalalignment='right');
industry_count.set_title('Prevalence of Various Industries Among Enforcement Actions')
```

```python
plt.figure(figsize=(16,9))

industry_fees = sns.boxplot(x="Industry", y="Total Assessed", data=df_enforcement)
industry_fees.set_xticklabels(industry_fees.get_xticklabels(), rotation=60, horizontalalignment='right');
industry_fees.set_title('Distribution of Fees Across Various Industries')
```

 Summary and Analysis
 
 We found that the industries which had the greatest number of cases of enforcement action being taken were Construction and Restaurant/Hotel. The next closest industry has less than half as many cases as either of these two. 
 
 The "Total Assessed" in the second graph is the sum of the restitution and penalty when enforcement action is taken against an employer. We see from the graph that across all industries, this is generally low, however there is a small number of cases where this total is exceptionally high. It appears that the Bank/Financial industry has the highest 75th quartile, with the next being Domestic Work. Also, it seems that the highest total assessment was in the Hospitals/Nursing Homes/Healthcare industry.
